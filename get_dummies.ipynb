```python
import pandas as pd

# 1. Create a dataset with words
data = {
    'Friend': ['Krish', 'Friend2', 'Friend3', 'Friend4'],
    'Preference': ['Mountains', 'Beaches', 'Mountains', 'City']
}
df = pd.DataFrame(data)

# 2. Use get_dummies to turn words into math
# This creates a '1' if the person likes that category
df_encoded = pd.get_dummies(df, columns=['Preference'])

print("--- ORIGINAL DATA ---")
print(df)
print("\n--- MATH DATA (One-Hot Encoded) ---")
print(df_encoded)
```

```python
# 'drop_first=True' removes the first category column to save space/logic
df_final = pd.get_dummies(df, columns=['Preference'], drop_first=True)
print(df_final)
```
