```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 1. Imagine this is data from a real tech company
# Salary is in thousands, but Years of Experience is small. 
# AI models struggle when scales are this different.
raw_data = {
    'Monthly_Salary': [45000, 120000, 32000, 85000, 60000],
    'Years_Experience': [1, 12, 0, 6, 3],
    'Projects_Completed': [5, 45, 1, 20, 12]
}

df = pd.DataFrame(raw_data)

# 2. Initialize the MinMaxScaler
# This squashes every value to be between 0 and 1.
scaler = MinMaxScaler()

# 3. Apply the transformation
scaled_values = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled_values, columns=df.columns)

print("--- RAW CAREER DATA ---")
print(df)
print("\n--- PROFESSIONALLY SCALED DATA (Ready for AI) ---")
print(df_scaled)
```
