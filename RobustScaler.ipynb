```python
# AI Career Track 2026: Lesson 3
## Topic: Robust Scaling & Outlier Detection
#**Objective:** Handle extreme numerical values using Scikit-Learn to ensure model stability.
from sklearn.preprocessing import RobustScaler
import pandas as pd

# 1. Create data with a 'Crazy' Outlier
# Notice the last salary is massive compared to others
data = {
    'Salary': [45000, 55000, 60000, 48000, 5000000], 
    'Experience': [1, 2, 3, 1, 20]
}
df = pd.DataFrame(data)

# 2. Use the RobustScaler
# It uses the Median and Interquartile Range (IQR) 
# instead of the Max, so it isn't 'distracted' by outliers.
robust_scaler = RobustScaler()
df_robust = pd.DataFrame(robust_scaler.fit_transform(df), columns=df.columns)

print("--- SCALED DATA (WITH OUTLIER HANDLED) ---")
print(df_robust)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check for outliers visually
sns.boxplot(x=df['Salary'])
plt.title("Visualizing the Outlier")
plt.show()
```
